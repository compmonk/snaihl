```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os
```

```python
data_dir = "data"
domain = "https://www.ncbi.nlm.nih.gov"
if not os.path.exists(data_dir): os.makedirs(data_dir)
```

```python
def download_and_save(url):
    print("Scraping", url)
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    page_name = soup.title.text
    
    dir_to_save = os.path.join(data_dir, page_name)
    if not os.path.exists(dir_to_save): os.makedirs(dir_to_save)
    
    file_path = os.path.join(dir_to_save, "page.html")
    with open(file_path, "wb") as pdf_file:
        pdf_file.write(response.content)
        print("HTML Saved at", file_path)
    
    pdf_link = soup.find("a", string="PDF version of this title")
    if pdf_link is None:
        print("Cannot get PDF Link")
        return
    
    pdf_link = domain + pdf_link.get("href")
    
    response = requests.get(pdf_link)  
    file_path = os.path.join(dir_to_save, "page.pdf")
    with open(file_path, "wb") as pdf_file:
        pdf_file.write(response.content)
        print("PDF Saved at", file_path)
```

```python
rss_list = "https://www.ncbi.nlm.nih.gov/books/new/?report=rss"
response = requests.get(rss_list)
rss_soup = BeautifulSoup(response.text, "xml")
```

```python
for book in rss_soup.find_all("item"):    
    print(book.title.text, book.link.text)
    download_and_save(book.link.text)
```
